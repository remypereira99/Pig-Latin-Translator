```python
#Pig Latin Translator
```

```python
def pig_latin_translator():
    
    phrase = input('Please enter the word/phrase to be translated into pig latin:')
    
    phrase_list = list(phrase)
    
    vowels = ['a','e','i','o','u']
    
    if len(phrase.split()) == 1:
    
        if phrase_list[0] not in vowels:

            first_letter = phrase_list[0]

            phrase_list.append(first_letter)

            phrase_list.pop(0)

            phrase_list.append('ay')

            print(''.join(phrase_list))

        else:

            phrase_list.append('ay')

            print(''.join(phrase_list))
            
    else:
        
        full_phrase = []
        
        for word in phrase.split():
            
            word_list = list(word)
            
            if word_list[0] not in vowels:

                first_letter = word_list[0]

                word_list.append(first_letter)

                word_list.pop(0)

                word_list.append('ay')
            
            else:

                word_list.append('ay')
                
            full_phrase.append(''.join(word_list))
        
        print(' '.join(full_phrase))
                
```

```python
pig_latin_translator()
```
